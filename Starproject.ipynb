```python
from math import sqrt



import numpy as np

from astropy.io import fits

from astropy.utils.data import download_file

import matplotlib.pyplot as plt



def star_count(image, threshold):

    n_stars = 0

    star_list = []
    star_list_ratio = []
    for x in range(5, image.shape[0]-15, 1):

        for y in range(5, image.shape[1]-15, 1):

            star = image[x-2:x+2, y-2:y+2]

            L = image[x-7:x-2, y-2:y+7]

            R = image[x+2:x+7, y-7:y+3]

            U = image[x-2:x+7, y+2:y+7]

            D = image[x-7:x+2, y-7:y-2]

            

            star_total = np.sum(star)

            L_total = np.sum(L)

            R_total = np.sum(R)

            U_total = np.sum(U)

            D_total = np.sum(D)

            

            total = L_total+R_total+U_total+D_total

            total_pixels = (len(L.flatten())

                            + len(R.flatten())

                            + len(U.flatten())

                            + len(D.flatten()))

            star_mean = star_total/len(star.flatten())

            back_mean = total/total_pixels

            difference = star_mean-back_mean
            fgbg_ratio = star_mean / back_mean
            found_diff = (difference >= threshold)
            ratio_threshold = 1.7
            found_ratio = (fgbg_ratio >= ratio_threshold)
            if found_diff or found_ratio:
                print(f'FOUND: diff: {found_diff} - {difference}   ratio: {found_ratio} - {fgbg_ratio}')
            if found_diff:
                n_stars += 1
                if star_is_unique(star_list, (x, y), 8):
                    print('CANDIDATE:', n_stars,
                          x, y, star_mean, back_mean, difference)
                    print('    CONFIRMED')
                    star_list.append((x, y))
                else:
                    # print('    DUPLICATE')
                    pass
            if found_ratio:
                if star_is_unique(star_list, (x, y), 8):
                    star_list_ratio.append((x, y))

    return star_list, star_list_ratio





def star_is_unique(star_list, center, avoidance_r):

    """Scans through the list of already-found stars to see if this candidate

    is so close that it's really a duplicate.  If it's a duplicate return

    False, otherwise return True.

    """
    for star in star_list:
        if sqrt((star[0] - center[0])**2 + (star[1] - center[1])**2) <= avoidance_r:

            return False

    return True



# now test it

image_file = download_file('http://data.astropy.org/tutorials/FITS-images/HorseHead.fits', cache=True )

hdu_list = fits.open(image_file)

my_img = hdu_list[0].data

star_list, star_list_ratio = star_count(my_img, 4000)

print(star_list)

print('n_stars:', len(star_list), len(star_list_ratio))

xvals = [star[0] for star in star_list]
yvals = [star[1] for star in star_list]
xvalsr = [star[0] for star in star_list_ratio]
yvalsr = [star[1] for star in star_list_ratio]
plt.axes().set_aspect('equal')
print(xvalsr, yvalsr)
plt.scatter(xvals, yvals, color='blue', alpha=0.5)
plt.scatter(xvalsr, yvalsr, color='red', alpha=0.5)

plt.show()
```
